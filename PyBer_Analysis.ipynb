```python
#Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os as os
import statistics
import matplotlib.ticker as tlt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#Create and show the chart
plt.plot(x_axis,y_axis)
plt.xlabel("Months")
plt.ylabel("US Dollars")
plt.title("Total fare accumulated for each month")
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis,x_axis, c="red")
plt.legend("Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,50)
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis,y_axis,s=y_axis)
```

```python
y_axis_larger = []
for dot in y_axis:
    y_axis_larger.append(dot*3)
plt.scatter(x_axis,y_axis,s=y_axis_larger)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=[i*5 for i in y_axis],c="skyblue", alpha = 0.8, edgecolors="black")
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis],c="skyblue", alpha = 0.8, edgecolors="black",label="Boston")
ax.set_title("PyBer Fare By Month")
ax.set_xlim(0,50)
ax.legend("Boston")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")
ax = plt.gca()
ax.set_ylim(ax.get_ylim()[::-1])
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev

```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev)
```
